# 1. Continuous Variable Exploration

```python
contin_feat = list(train.select_dtypes(include=[np.number]))
contin_feat.remove("Id")
contin_feat.remove("MSSubClass")
```

```python
plt.figure(figsize=(20,15))
ix = train[contin_feat].corr().sort_values("SalePrice", ascending = False).index
df_sorted_by_correlation = train[contin_feat].loc[:, ix]
fifteen_more_correlated = df_sorted_by_correlation.iloc[:, :15]
corr = fifteen_more_correlated.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    # display a correlation heatmap
    ax = sns.heatmap(corr, mask=mask, annot=True)
```

```python
plt.figure(figsize=(20,15))
sns.boxplot(x="OverallQual", y="SalePrice", data=train[['OverallQual', 'SalePrice']])
plt.show()
```
